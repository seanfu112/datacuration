###Imported Necessary APIs and Modules
- Imported Pandas, MatPlotLib, and BeautifulSoup

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
from bs4 import BeautifulSoup
```

###Imported List of Countries and its population & percentage & Date Updated using Pandas
- Set Countries as index
- Calculate percentage by dividing country population by world population


```python
url = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population"
infotable = pd.read_html(url, match = "Country", flavor = 'bs4')
infotable = infotable[0]
df = pd.DataFrame()
df['Country'] = infotable['Country / Dependency']
df['Population'] = infotable['Population']
df['Percentage of World'] = df["Population"]/8063792000
df["Date Updated"] = infotable["Date"]
df = df.drop([0],axis=0)
df = df.set_index("Country")
df
```

###Use MatPlotLib to Draw a Bar graph for the 30 countries with the largest population.

```python
temp = df[0:30]
fig = plt.figure(figsize=(20,30))
ax=fig.add_subplot(311)
ax.bar(temp.index, temp["Population"])
ax.set(title="Population of 30 Largest Countries", xlabel='Countries', ylabel='Population')
ax.set_xticklabels(temp.index,rotation=90)
# ax.yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0))
```

###Use MatPlotLib to draw a pie chart and distribution for the top 30 countries with the largest populations.

```python
plot=temp.plot.pie(y='Population', title = "Percentage of Population for 30 Largest Countries",figsize=(20, 10))
```

### Export Data in the DataFrame to a CSV file

```python
df.to_csv("Data.csv")
```
